# Tokenization

The first step of processing we make the texts go through is to split the raw sentences into words, or more exactly tokens. The easiest way to do this would be to split the string on spaces, but we can be smarter:

- we need to take care of punctuation
- some words are contractions of two different words, like isn't or don't

To see what the tokenizer had done behind the scenes, let's have a look at a few texts in a batch.

````python
# import nltk
# nltk.download('stopwords')

# from nltk.corpus import stopwords 
# # stop_words
# stop_words = stopwords.words('english')
# stop_words
from google.colab import files
files.upload()
````

````python
stop_words=pd.read_excel('hindi stopwords.xlsx')
print (str(stop_words))
````

````python
# detokenized_docs=[]
# t = ' '.join(tokenized_doc[1]) 
# detokenized_docs.append(t) 
# detokenized_docs
````

````python
# tokenization 
tokenized_doc = df['Query'].apply(lambda x: x.split())
# tokenized_doc

````

````python
# remove stop-words 
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in str(stop_words)])
tokenized_doc
````

````python
# de-tokenization 
detokenized_doc = [] 
for i in range(len(df)): 
    t = ' '.join(tokenized_doc[i]) 
    detokenized_doc.append(t) 

df['Query'] = detokenized_doc
````

````python
df
````

### split the dataset into train & validation sets

````python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
````

````python
df1=train.reset_index()
df2=test.reset_index()
````

````python
train_df=df1.drop('index',axis=1)
valid_df=df2.drop('index', axis=1)
````

````python
path = '/content'
````

## language model

````python
# Language model data
data_lm = TextLMDataBunch.from_df(path,train_df , valid_df )

# Classifier model data
data_clas = TextClasDataBunch.from_df(path,train_df,valid_df , vocab=data_lm.train_ds.vocab, bs=32)
````

# Fine-Tuning the Pre-Trained Model and Making Predictions

````python
learn = language_model_learner(data_lm,arch=AWD_LSTM,pretrained=False, drop_mult=0.3)
````

````python
# train the learner object with learning rate = 1e-2
learn.fit_one_cycle(5, 1e-2)
````

````python
learn.save_encoder('ft_enc')
````

````python
learn = text_classifier_learner(data_clas,arch=AWD_LSTM,pretrained=False, drop_mult=0.3)
learn.load_encoder('ft_enc')
````

````python
learn.fit_one_cycle(5, 1e-2)
````

## predictions

````python
# get predictions
preds, targets = learn.get_preds()

predictions = np.argmax(preds, axis = 1)
pd.crosstab(predictions, targets)
````
